### Beginner Question 1:

Using the [SCC dataset](https://huggingface.co/datasets/refugee-law-lab/canadian-legal-data) from the Refugee Law Lab, print the names of the five longest Supreme Court of Canada cases.

```python
#!pip install datasets
#!pip install pandas

from datasets import load_dataset
dataset = load_dataset("refugee-law-lab/canadian-legal-data", split="train", data_dir="SCC")

import pandas as pd
df = pd.DataFrame(dataset)
```

```python
# Assuming that the question is asking for the length of the text of the case in characters

# create colum for the length of the text of the case
df['case_len'] = df["unofficial_text"].apply(len)

# print df['name'] for 5 highest df['case_len'] (print as strings)
for i in df.nlargest(5, 'case_len')['name']:
    print(i)
```

```python
# Assuming that the qustion is asking for the length of the case name in characters

# create column for the length of the case name
df['name_len'] = df["name"].apply(len)

# print df['name'] for 5 highest df['name_len'] (print as strings)
for i in df.nlargest(5, 'name_len')['name']:
    print(i)
    print()
```

```python
# another option assuming that the question is asking for the length of the case name in characters
# credit to Athaven Nithianantha for a version of this solution

df.sort_values(by="name", key=lambda x: x.str.len(), ascending=False).head(5).reset_index(drop=True)
```

```python
# Assuming that the question is asking for the length of the text of the case in words
# using a custom function

# count the number of words in the text by splitting the text by spaces
def word_count(text):
    return len(text.split())

# create column for the length of the text of the case in words
df['case_words'] = df["unofficial_text"].apply(word_count)

# print df['name'] for 5 highest df['case_words'] (print as strings)
for i in df.nlargest(5, 'case_words')['name']:
    print(i)
```

```python
# Assuming that the question is asking for the length of the text of the case in words
# using lamda function

# create column for the length of the text of the case in words
df['case_words'] = df["unofficial_text"].apply(lambda x: len(x.split()))

# print df['name'] for 5 highest df['case_words'] (print as strings)
for i in df.nlargest(5, 'case_words')['name']:
    print(i)
```

```python
# Alternative ways of locating largest values in a column

# using nlargest
for i in df.nlargest(5, 'case_len')['name']:
    print(i)

print('_____________')

# using sort_values
for i in df.sort_values('case_len', ascending=False).head(5)['name']:
    print(i)

print('_____________')

# not printing as strings
print(df.nlargest(5, 'case_len')['name'])
```

```python
# Another solution, credit to Kristen Wong for a variation of this solution

judges_url = 'https://raw.githubusercontent.com/srehaag/legal_info_tech_w24/main/judges.json'
df = pd.read_json(judges_url)
df

# prompt: identify whether each entry under column "bio" uses predominantly she/her/Ms./Mrs. or he/him/Mr. gender pronouns

import re

def identify_gender(bio):
  """
  Identifies the predominant gender pronouns used in a biography.

  Args:
    bio: A string containing the biography.

  Returns:
    "female" if she/her/Ms./Mrs. pronouns are used predominantly,
    "male" if he/him/Mr. pronouns are used predominantly,
    "unknown" otherwise.
  """

  she_count = len(re.findall(r'\bmadam\b|\bshe\b|\bher\b|Ms\.?|\bMrs\.?\b', bio, flags=re.IGNORECASE))
  he_count = len(re.findall(r'\bhe\b|\bhim\b|Mr\.?\b', bio, flags=re.IGNORECASE))

  if she_count > he_count:
    return "female"
  elif he_count > she_count:
    return "male"
  else:
    return "unknown"

# Apply the function to the "bio" column
df["gender"] = df["bio"].apply(identify_gender)

# Count the number of female and male judges
female_count = df["gender"].value_counts().get("female", 0)
male_count = df["gender"].value_counts().get("male", 0)

# Calculate the proportion of female and male judges
total_judges = len(df)
proportion_female = female_count / total_judges
proportion_male = male_count / total_judges

# Print the results
print(f"Proportion of female judges: {proportion_female:.2f}")
print(f"Proportion of male judges: {proportion_male:.2f}")
```

```python
dataset = load_dataset("refugee-law-lab/canadian-legal-data", split="train", data_dir="RAD")
df = pd.DataFrame(dataset)

# Find all fc citations using regex
import re
fc_regex = re.compile(r'\b\d{4}\sFC\s\d{1,4}\b')
fc_cites = fc_regex.findall(" ".join(df['unofficial_text']))

# list the 10 most common fc cites using Counter
from collections import Counter
Counter(fc_cites).most_common(10)
```

```python
# Another approach: Credit to Brandon Pierre for a version of this solution
from datasets import load_dataset
import re

def search_string(pattern, text):
    matches = []
    for t in text:
        matches += [x.group() for x in re.finditer(pattern, t)]
    return pd.DataFrame(matches, columns=['Neutral Citation']).value_counts().reset_index(name='Frequency')

df = pd.DataFrame(load_dataset("refugee-law-lab/canadian-legal-data", split="train", data_dir= "RAD"))

#rRegex pattern - YYYY FC #
pattern = r'(18|19|20)\d{2}(\sFC\s)\d+'

#Calculate frequency of search term -- format YYYY FC #
text = df.unofficial_text.to_list()

#Print top 10 cases FC cited in RAD
display(search_string(pattern,text).head(10))
```

```python
# And another approach (credit to Denis Keimakh for a version of this solution)

import pandas as pd
import re
from collections import Counter

# Load the dataset
df = pd.DataFrame(load_dataset("refugee-law-lab/canadian-legal-data", split="train", data_dir= "RAD"))

# Pattern to match Federal Court of Canada case citations. Looks for strings that start and end at word boundaries, with a format that includes a four-digit year, followed by " FC " and the case number.
citation_pattern = r"\b(\d{4} FC \d+)\b"

# Extract citations from the "unofficial_text" column and count occurrences
tracked_citations = Counter()
for fulltext in df["unofficial_text"]:
    # Find all matches of the pattern in the text
    citations = re.findall(citation_pattern, fulltext)
    # Update counts for each citation found
    tracked_citations.update(citations)

# Get the top 10 most common citations
top_10_citations = tracked_citations.most_common(10)

# Print the top 10 citations and their counts
for citation, count in top_10_citations:
    print(f"{citation}: {count}")
```

### Intermediate Question 2:

Part 1: Calculate the Flesch Reading Ease score for two datasets in the Refugee Law Lab's [Canadian Legal Data](https://huggingface.co/datasets/refugee-law-lab/canadian-legal-data) datasets of your choice (e.g. Tax Court of Canada vs Canadian Human Rights Tribunal) for decisions in the past year.

Part 2: Create line charts for the average Flesch Reading Ease score per year for each of the two tribunals.

Hint: Search for a python package that calculates Flesch Reading Ease, rather than calculating yourself.

```python
import pandas as pd
from datasets import load_dataset

dfFCA = pd.DataFrame(load_dataset("refugee-law-lab/canadian-legal-data", split="train", data_dir="FCA"))
dfFC = pd.DataFrame(load_dataset("refugee-law-lab/canadian-legal-data", split="train", data_dir="FC"))
dfFCA = dfFCA[dfFCA['language'] == 'en']
dfFC = dfFC[dfFC['language'] == 'en']

#!pip install textstat
import textstat

def flesch(text):
    return textstat.flesch_reading_ease(text)

dfFCA['flesch'] = dfFCA['unofficial_text'].apply(flesch)
dfFC['flesch'] = dfFC['unofficial_text'].apply(flesch)

print('Average Flesch Reading Ease for FCA in 2023:', round(dfFCA[dfFCA['year'] == 2023]['flesch'].mean()))
print('Average Flesch Reading Ease for FC in 2023:', round(dfFC[dfFC['year'] == 2023]['flesch'].mean()))
```

```python
# line chart of average flesch reading ease by year: FCA
dfFCA.groupby('year')['flesch'].mean().plot()
```

```python
# line chart of average flesch reading ease by year: FC
dfFC.groupby('year')['flesch'].mean().plot()
```

Ask the user to input their choice of Refugee Law Lab's [Canadian Legal Data](https://huggingface.co/datasets/refugee-law-lab/canadian-legal-data) datasets (e.g. SCC, TCC, FC, CHRT, etc). Then ask the user for a word or phrase for which they would like to see the frequency of use in that dataset over time. Then based on the user's input create a line chart showing three year rolling averages for the number of times that word appears in the selected dataset.

For an online example of implementation of something similar, see:

```python
# Load full dataset (requires a reasonable amount of RAM)
df = pd.DataFrame(load_dataset("refugee-law-lab/canadian-legal-data", split="train"))

# drop rows where no df.year
df = df.dropna(subset=['year'])

# convert year to datetime but with year only
df['year'] = pd.to_datetime(df['year'], format='%Y').dt.year
```

```python
# ask user for dataset
selected_dataset = input("Which dataset would you like to use? (e.g. SCC) ").upper()

# filter dataset for selected dataset
dfSelected = df[df['dataset'] == selected_dataset]

# filter dataset for language = 'en' (optional)
dfSelected = dfSelected[dfSelected['language'] == 'en']

# ask user for the word or phrase to search for
search_term = input("What word or phrase would you like to search for? ")

# count the number of times the search_term appears in the unofficial_text
dfSelected['search_term_count'] = dfSelected['unofficial_text'].str.lower().str.count(search_term.lower())

# graph 3 year rolling average
dfSelected.groupby('year')['search_term_count'].sum().rolling(3).mean().plot(title=f"\"{search_term}\" occurrences, 3 year rolling average in {selected_dataset} decisions")
```
