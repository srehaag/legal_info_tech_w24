### Beginner Question 2:

Find a course summary (yours or one that you find online) for an Osgoode Hall Law School course in MS Word docx format.

Extract the text of the document into a variable called: course_summary_text

Print the first 500 characters of the course_summary_text variable

```python
# Using python-docx  (uncomment pip install if needed)

# ! pip install python-docx

from docx import Document
doc = Document("my_course_summary.docx")

text = ""
for para in doc.paragraphs:
    text = text + para.text + "\n"
print(text[0:500])
```

```python
# Alternatively, use +=

# ! pip install python-docx

from docx import Document
doc = Document("my_course_summary.docx")

text = ""
for para in doc.paragraphs:
    text += para.text + "\n"
print(text[0:500])
```

### Intermediate Question 1:

Using the same dataframe about refugee claims in 2019 from Beginner Question 3, filter the dataframe so that it only includes positive and negative decisions (excluding claims that were otherwise resolved).

Positive decisions include the following:


*   Positive
*   Expedited Positive

Negative decisions include the following:

*   Negative
*   Neg. No Cred Basis

Using the filtered dataframe, caculate and print the recognition rate (proportion of positive relative to the sum of positive and negative) for the twenty highest volume lawyers.

```python
# setup
import pandas as pd

# load data
url = "https://refugeelab.ca/wp-content/uploads/2024/01/2019_RPD_Data.xlsx"
df = pd.read_excel(url)

# filter for cases with positive or negative outcomes
df = df[df["Explanation"].isin(["Positive","Expedited Positive", "Negative", "Neg. No Cred Basis"])]

# add column with outcome (positive = True, negative = False)
df["outcome"] = df["Explanation"].isin(["Positive","Expedited Positive"])

# Get list of top 20 counsel
top20 = df["Counsel Fullname"].value_counts().head(20).index.to_list()

# Filter df for top 20 counsel
df = df[df["Counsel Fullname"].isin(top20)]

# get proportion of outcome True for each counsel
df.groupby("Counsel Fullname")["outcome"].mean().sort_values(ascending=False)
```

Using one or the other, put the text of the  *Genetic Non-Discrimination Act* into a variable called target_statute, and print that variable.

```python
# using individual link
# ! pip install bs4

import requests
from bs4 import BeautifulSoup

url = "http://laws-lois.justice.gc.ca/eng/XML/G-2.5.xml"
response = requests.get(url)

# Parse the XML document with BeautifulSoup
soup = BeautifulSoup(response.content, 'lxml')

# Print out the text of the XML document without the tags
print(soup.get_text("\n", strip=True))
```

```python
# Using full XML
import requests
from lxml import etree
from bs4 import BeautifulSoup

url = 'https://laws-lois.justice.gc.ca/eng/XML/Legis.xml'
response = requests.get(url).content

# Parse the XML content to find URL for statute
root = etree.fromstring(response)
law = "Genetic Non-Discrimination Act"
match = root.xpath(f"//Title[text()='{law}']")[0]
url = match.getparent().find('LinkToXML').text

# get the XML for the target statute
response = requests.get(url)
soup = BeautifulSoup(response.content, 'lxml')
print(soup.get_text("\n", strip=True))


# NOTE: XML matching process by Brandon Pierre who came up with a better
# way to find statutes in full XML than I did
```

### Advanced Question 1:

Scrape the biographies of all judges on the Supreme Court of Canada [website](https://www.scc-csc.ca/judges-juges/cfpju-jupp-eng.aspx), and put them into a dataframe that includes the name of the judge, the date of appointment, the date of departure (if applicable), the biography, and the URL where the biography can be found.

Some hints regarding webscraping in Python:

*   Use requests (requests.get(url).content) to get html
*   Use a python package (such as beautifulsoup) to parse the html


```python
# ! pip install bs4

# setup
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# get url
url = "https://www.scc-csc.ca/judges-juges/cfpju-jupp-eng.aspx"
response = requests.get(url)
print(response.text)
```

```python
# check for tables in html and put into a list
soup = BeautifulSoup(response.content, 'html.parser')
tables = soup.find_all("table")

# only one table, so we can use index 0
table = tables[0]

# convert table to df
df_judges = pd.read_html(str(table))[0]
df_judges
```

```python
# get all links from the table
links = table.find_all("a")
for link in links:
    print(link.get("href"))
```

```python
# clean up links

root_url = "https://www.scc-csc.ca/judges-juges/"

bio_links = []

for link in links:
    # if link does not start with #
    if not link.get("href").startswith("#"):
        # add root_url to link
        bio_link = root_url + link.get("href")
        # add to list
        bio_links.append(bio_link)

# add bio_links to df_judges
df_judges["bio_link"] = bio_links
df_judges
```

```python
# try to get one bio
url = df_judges["bio_link"][0]
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# using web developper tools in browser to inspect the HTML, can see that text of
# bio is in div id = "wb-main", so get the text from that

bio = soup.find("div", {"id": "wb-main"}).get_text(" ", strip=True)
```

```python
# The bio has some additional information (title duplicated, alt-text
# for images, modified date, etc.). Let's clean it up

# get the bio again from prior cell, but use a custom splitter between paras
bio = soup.find("div", {"id": "wb-main"}).get_text("|", strip=True)
print(bio)

# split on custom splitter
paras = bio.split("|")

# remove first two para and last two para, and then join
bio = " ".join(paras[2:-2])
print(bio)

# Looks good!
```

```python
# get all the bios using above code

# set up a function from the above cells
def get_bio(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    bio = soup.find("div", {"id": "wb-main"}).get_text("|", strip=True)
    paras = bio.split("|")
    bio = " ".join(paras[2:-2])
    # add 1 second delay to be nice to the SCC servers
    time.sleep(1)
    return bio

# apply function to df_judges (should take about 2 minutes)
df_judges["bio"] = df_judges["bio_link"].apply(get_bio)
```
